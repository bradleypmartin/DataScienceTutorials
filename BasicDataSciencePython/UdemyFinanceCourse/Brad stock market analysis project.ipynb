```python
### In this project, we're going to do some elementary financial analysis on the stocks of several auto companies. ###

# importing necessary libraries
import pandas_datareader.data as web
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
# setting start of analysis at Jan 1, 2012; running to Jan 1, 2017.
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2017,1,1)
```

```python
# grabbing data for $TSLA (note: this is dependent on Quandl's datasets being available on your network)
tesla = web.DataReader('TSLA','quandl',start,end)
```

```python
# checking out a cross section of $TSLA's data
tesla.tail()
```

```python
# grabbing data for $F (Ford)
ford = web.DataReader('F','quandl',start,end)
ford.tail()
```

```python
# grabbing data for $GM
GM = web.DataReader('GM','quandl',start,end)
GM.tail()
```

```python
# Next, we'll start visualizing the data.  Below is a plot of the market open prices for the three automakers
# above ($TSLA, $F, $GM) between 2012 and 2017.
tesla['Open'].plot(figsize=(16,6))
GM['Open'].plot()
ford['Open'].plot()
plt.legend(['Tesla Open','GM Open','Ford Open']);
```

```python
# Daily volume is another really important metric to check out; we'll plot that next.
tesla['Volume'].plot(figsize=(16,6))
GM['Volume'].plot()
ford['Volume'].plot()
plt.legend(['Tesla vol.','GM vol.','Ford vol.'])
```

** Interesting, looks like Ford had a really big spike somewhere in late 2013. What was the date of this maximum trading volume for Ford?**

**Bonus: What happened that day? **

```python
# Looking at the end of 2013, we see a large volume spike that coincides with a price trough. We can find the timestamp
# of this volume spike like so:

ford['Volume'].idxmax()

# What gives? A quick Google search shows that Ford beat earnings expectations during this time,
# but also forecast possible losses in Europe and plans to roll out one or more new vehicles.
```

```python
# The next thing we're going to look at is Total Traded (TT) daily dollar volume ((open price)*(daily volume)).
# We're not accessing market cap here, but perhaps this is a "good enough" local surrogate for now.
tesla['Total Traded'] = tesla['Open'].multiply(tesla['Volume'])
GM['Total Traded'] = GM['Open'].multiply(GM['Volume'])
ford['Total Traded'] = ford['Open'].multiply(ford['Volume'])
tesla['Total Traded'].plot(figsize=(16,6))
GM['Total Traded'].plot()
ford['Total Traded'].plot()
plt.legend(['Tesla TT.','GM TT.','Ford TT.']);
```

```python
# As with Ford and the last data set, we can quickly find a timestamp for the crazy spike in early 2014 and then do a
# Google search to find out what happened at that time to create so much activity in Tesla stock trading.

tesla['Total Traded'].idxmax()

# It looks like a strong earnings report the week before, coupled with rumors of the Gigafactory plan and
# great reception of the Model S ('Car of the Year' was given to it by Consumer Reports that day, apparently) drove trading.
```

```python
# Moving averages are important technical analysis tools for equity traders.  Let's make sure our $GM data is in the correct
# order for plotting (increasing dates) and take a look at some $GM moving averages.

GM = GM.sort_index(axis=0,ascending=True)
GM.head()
```

```python
GM['MA50'] = GM['Close'].rolling(window=50).mean()
GM['MA200'] = GM['Close'].rolling(window=200).mean()
GM[['Open','MA50','MA200']].plot(figsize=(16,6));
```

```python
# Next, we're going to look graphically for correlations between the three automakers' stock prices.
# to do this, we're going to use a scatter matrix.

from pandas.plotting import scatter_matrix as sm
```

```python
opendf = pd.DataFrame(tesla['Open'])
newdf = opendf.rename(index=str, columns={"Open":"Tesla Open"})
opendf = newdf
opendf['Ford Open'] = ford['Open']
GM = GM.sort_index(axis=0,ascending=False)
opendf['GM Open'] = GM['Open']
sm(opendf, alpha = 0.2, figsize=(6,6),diagonal='hist');

# From the results, it appears GM and Ford have a strong positive correlation (no surprise here) but that
# their correlation with Tesla's price is much less apparent.  Note on the diagonal we've got histograms of opening
# prices for all three automakers' stock; Tesla's meteoric rise is really apparent here!
```

```python
# Another really popular plot for traders to visualize is a candlestick chart (showing high, low, beginning, and ending stock
# prices during a designated window of time). We'll use some specialized functions just for this purpose.

newstart = datetime.datetime(2014,1,1)
newend = datetime.datetime(2014,1,31)

import matplotlib.ticker as mticker
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
newford = web.DataReader('F','quandl',newstart,newend)
```

```python
newford = newford.sort_index(axis=0,ascending=True)
# forming and limiting new df to data needed by candlestick plot
newford_temp = newford.reset_index()
newford_temp = pd.DataFrame(newford_temp[['Date','Open','High','Low','Close']])

# converting dates
newford_temp['Date'] = newford_temp['Date'].map(mdates.date2num)

# making the plot
fig = plt.figure()
ax = plt.subplot2grid((6,1),(0,0),rowspan=6,colspan=1)

# converting raw mdates to dates
ax.xaxis_date()
plt.xlabel('Date')
print(newford_temp)

# making candlestick
candlestick_ohlc(ax,newford_temp.values,width=0.5,colorup='g',colordown='r',alpha=0.75)
plt.ylabel('Price')
#plt.legend()

plt.show()
```

```python
# The next thing we're going to do is look at daily percentage change in each of the equities.
# Before doing this, we'll cull our data frames a bit.

newGM = pd.DataFrame(GM[['Open','High','Low','Volume','Close','Total Traded']])
newGM = newGM.sort_index(axis=0,ascending=True)
newGM['returns']=newGM['Close'].pct_change(periods=1)
#newGM.head()
newTesla = pd.DataFrame(tesla[['Open','High','Low','Volume','Close','Total Traded']])
newTesla = newTesla.sort_index(axis=0,ascending=True)
newTesla['returns']=newTesla['Close'].pct_change(periods=1)

newFord = pd.DataFrame(ford[['Open','High','Low','Volume','Close','Total Traded']])
newFord = newFord.sort_index(axis=0,ascending=True)
newFord['returns']=newFord['Close'].pct_change(periods=1)
```

```python
newGM.head()
```

```python
# Here's a histogram of Ford's daily % returns, for example...
newFord['returns'].hist(bins=50);
```

```python
# ...and here's one for GM...
newGM['returns'].hist(bins=50);
```

```python
# ...and also for Tesla. Compared to the other two, I'd say it's fair to guess that Tesla is definitely the most
# volatile stock (high variance displayed here)
newTesla['returns'].hist(bins=50);
```

```python
# If we wish, we can make one plot to handle the histograms for all three automakers' daily percent change.

fig = plt.figure(figsize=(10,6))
newTesla['returns'].hist(bins=100,alpha=0.5)
newGM['returns'].hist(bins=100,alpha=0.5)
newFord['returns'].hist(bins=100,alpha=0.5)
plt.legend(['Tesla','GM','Ford']);
```

```python
# We can also try smoothing the histograms into continuous variants with a Kernel Density Estimate (KDE) plot.

fig = plt.figure(figsize=(10,6))
newTesla['returns'].plot.kde()
newGM['returns'].plot.kde()
newFord['returns'].plot.kde()
plt.legend(['Tesla','GM','Ford'])
```

```python
# Yet another way to visualize these percent return distributions is through the use of box plots (mean, quartiles, outliers).

fig = plt.figure(figsize=(8,6))
returns_df = pd.DataFrame(newTesla['returns'])
returns_df = returns_df.rename(index=str, columns={"returns":"Tesla returns"})
returns_df['GM returns'] = newGM['returns']
returns_df['Ford returns'] = newFord['returns']
returns_df.boxplot();
```

```python
# As with price action, we can take a look at a scattermatrix of daily % returns to get an idea of possible
# correlations between the equities.

sm(returns_df, alpha = 0.2, figsize=(6,6),diagonal='hist');
```

```python
# isolating the GM/Ford relationship to get a better idea of the correlation there

returns_df.plot.scatter('GM returns','Ford returns',figsize=(10,8),alpha=0.5);
```

```python
# Maybe a more useful "at-a-glance" plot for potential long-term investors would be a plot of CUMULATIVE returns over time.
# We'll create one of those here by compounding daily returns that we've already computed.

newTesla['Cumulative Return'] = (1+newTesla['returns']).cumprod()
newFord['Cumulative Return'] = (1+newFord['returns']).cumprod()
newGM['Cumulative Return'] = (1+newGM['returns']).cumprod()
```

```python
# Here we're plotting the cumulative returns. Note that the vertical (y-) axis displays a multiplication factor for an initial
# investment made in early 2012.  Tesla investors from that period are probably quite satisfied with their ~8x capital gains!

fig = plt.figure(figsize=(16,8))
newTesla['Cumulative Return'].plot()
newGM['Cumulative Return'].plot()
newFord['Cumulative Return'].plot()
plt.legend(['Tesla','GM','Ford']);
```
